### Visualización de Datos

```python
# distribución de generos

plt.figure(figsize=(12,6))

sns.countplot(x= "Genre", data=df)

# TODO: add plt title
plt.title("Distribución de Genero")

plt.xlabel("Genero de Peliculas")
plt.ylabel("Numero de Peliculas por Genero")

plt.xticks(rotation= "vertical")
plt.show()
```

### Grafica con otra categoria distribución por lenguaje

```python
plt.figure(figsize=(12,6))

sns.countplot(x= "Original language", data=df, order= df["Original language"].value_counts().index)

# TODO: add plt title
plt.title("Distribución de Idiomas")

plt.xlabel("Genero de Libros")
plt.ylabel("Numero de Idiomas")

plt.xticks(rotation= 50)
plt.show()
```

### Publicaciones en el Tiempo

```python
# Visualizar tipo de archivos

df.dtypes
```

```python
df['First published'] = pd.to_datetime(df['First published'], errors='coerce')
df.dtypes
df.head()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    x='First published',
    y=df.index,
    data=df
)

plt.title('Libros publicados en el tiempo')
plt.xlabel('año')
plt.ylabel('Indice')
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    x='First published',
    y=df.index,
    data=df.sort_values('First published')
)

plt.title('Libros publicados en el tiempo')
plt.xlabel('año')
plt.ylabel('Indice')
plt.show()
```

### Agrupar Información

```python
author_year_counts = df.groupby(
    ['Author(s)', df['First published'].dt.year]
).size().reset_index(name='Book count')

author_year_counts

```

```python
pivot_table= author_year_counts.pivot(index='First published', columns= "Author(s)", values= "Book count").fillna(0)
pivot_table
```

```python
import warnings
warnings.filterwarnings("ignore") 

ax = pivot_table.plot(kind='line', marker='o', figsize=(12,6))

plt.title('Numero de libros publicados por author en el tiempo')
plt.xlabel('año')
plt.ylabel('Libros')

ax.legend(
    title='Autor',
    loc='upper left',
    bbox_to_anchor=(1,1)
)

plt.show()
```
